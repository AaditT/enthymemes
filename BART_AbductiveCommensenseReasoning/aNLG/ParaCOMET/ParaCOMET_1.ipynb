```python
%cd content/
text_file = open("train.source", "r")
lines = text_file.readlines()
print(len(lines))
text_file.close()
new_lines = []
for line in lines:
  new_lines.append(line.strip())
new_lines[0]
```

```python
id_counter = 3001
json_lines = []
for line in new_lines:
  argument_list = line.split(" # ")

  json_instance = {
    "full_context": [
      argument_list[0],
      argument_list[1],
    ],
    "id":id_counter
  }
  if (id_counter == 1):
    print(json_instance)
  json_lines.append(json_instance)
  id_counter += 1
print(len(json_lines))
```

```python
import json

with open('example.jsonl', 'w') as outfile:
  for entry in json_lines:
    json.dump(entry, outfile)
    outfile.write('\n')
```
